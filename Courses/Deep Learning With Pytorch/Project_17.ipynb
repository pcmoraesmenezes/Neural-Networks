# Project 17 - Classify a Homer or Bart Simpsons with a Dense Neural Network

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import torch
from torch import nn, optim
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
torch.__version__
```

```python
df = pd.read_csv('Datasets/personagens.csv')

labels = df.iloc[:, 6].values
forecasters = df.iloc[:, 0:6].values
```

```python
forecasters
```

```python
le = LabelEncoder()
labels = le.fit_transform(labels)


labels
```

```python
train_X, test_X, train_y, test_y = train_test_split(forecasters, labels, test_size=0.3, random_state=42)
```

```python
train_X = torch.tensor(np.array(train_X), dtype=torch.float32)
train_y = torch.tensor(np.array(train_y), dtype=torch.float32)
```

```python
dataset = torch.utils.data.TensorDataset(train_X, train_y)
train_loader = torch.utils.data.DataLoader(dataset, batch_size=10, shuffle=True)
```

```python
class classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.dense0 = nn.Linear(6, 4)
        self.dropout = nn.Dropout(0.2)
        self.activation0 = nn.ReLU()
        self.dense1 = nn.Linear(4, 4)
        self.activation1 = nn.ReLU()
        self.dense2 = nn.Linear(4, 1)
        self.activation2 = nn.Sigmoid()

    def forward(self, X):
        X = self.dense0(X)
        X = self.dropout(X)
        X = self.activation0(X)
        X = self.dense1(X)
        X = self.activation1(X)
        X = self.dense2(X)
        X = self.activation2(X)
        return X
```

```python
model = classifier()
```

```python
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters())
```

```python
for epoch in range(2000):
    
    running_loss = 0.
    running_accuracy = 0.
    
    for data in train_loader:
        inputs, label = data
        
        label = label.view(-1, 1)

        optimizer.zero_grad()        

        outputs = model(inputs)
        loss = criterion(outputs, label)
        loss.backward()
        
        optimizer.step()

        outputs = outputs.flatten().detach().numpy()

        outputs = (outputs > 0.5).astype('int32')

        label = label.flatten().detach().numpy()

        accuracy = accuracy_score(label, outputs)

        running_loss += loss.item()
        running_accuracy += accuracy

    print(f'Epoch {epoch+1} - Loss: {running_loss/len(train_loader):.4f} - Accuracy: {running_accuracy/len(train_loader):.4f}')
```

```python
model.eval()
```

```python
X_test = torch.tensor(np.array(test_X), dtype=torch.float32)
y_test = torch.tensor(np.array(test_y), dtype=torch.float32)
```

```python
outputs = model(X_test)

loss = criterion(outputs, y_test.view(-1, 1))

label = y_test.flatten().detach().numpy()

outputs = outputs.flatten().detach().numpy()
outputs = (outputs > 0.5).astype('int32')

accuracy = accuracy_score(label, outputs)

print(f'Loss: {loss.item():.4f} - Accuracy: {accuracy:.4f}')
```
